```python
import pandas as pd
```

```python
# 合併所有dataframe
# periods = [30,60,90,180,270,360]
periods = [30,60,90,180,270,360]
total_scores = pd.DataFrame()

for period in periods:
    perios_socres = pd.read_parquet(f'params/params_SVD_stride30/surprise_SVD_train{period}.parquet')
    total_scores = pd.concat([total_scores,perios_socres],axis=0,ignore_index=True)
    
total_scores
```

```python
# 看最佳時間區間
total_scores.groupby(['train_period']).mean('map@k').sort_values('map@k',ascending=False)
```

```python
# 看最佳的參數組合
total_scores.groupby(['train_period','factors','iterations','regularization']).mean('map@k').sort_values('map@k',ascending=False).head(20)
```

```python

total_scores.groupby(['train_period','factors']).mean('map@k').sort_values('map@k',ascending=False).head(10)
```

```python
total_scores.groupby(['train_period','iterations']).mean('map@k').sort_values('map@k',ascending=False).head(10)
```

```python
clean_scores = total_scores[total_scores.rmse < 10]
clean_scores.groupby(['train_period']).mean('rmse').sort_values('rmse',ascending=False)
```
